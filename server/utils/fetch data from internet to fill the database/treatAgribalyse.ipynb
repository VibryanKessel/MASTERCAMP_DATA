```python
import pandas as pd
```

```python
df = pd.read_csv("agribalyse-detail-etape.csv")

df
```

```python
grp_df = df.groupby(by="Sous-groupe d'aliment").mean()
grp_df
```

```python
grp_df.columns
```

On est intéressé par les score EF des aliments 

```python
grp_df = grp_df[[ 'Score unique EF (mPt / kg de produit) - Agriculture',
       'Score unique EF (mPt / kg de produit) - Transformation',
       'Score unique EF (mPt / kg de produit) - Emballage',
       'Score unique EF (mPt / kg de produit) - Transport',
       'Score unique EF (mPt / kg de produit) - Supermarché et distribution',
        'Score unique EF (mPt / kg de produit) - Consommation']]
```

```python
grp_df.head(5)
```

L'idée pour le moment est de donner un score (eco-score) global à chaque sous-groupe d'aliments. Nous allons donc faire la moyenne des score pour chaque sous-groupe

```python
t = grp_df.to_numpy()
```

```python
grp_df = pd.DataFrame(index= grp_df.index, columns=["Eco-Score"],data=t.mean(axis=1))
```
